# PUBLICATION : Hydrogen Adsorption on Au Surfaces

This notebook analyzes Density Functional Theory (DFT) results for hydrogen adsorption on Au surfaces.  
Key data (e.g., adsorption energies, surface deformations) are extracted from VASP output files and visualized.


```python
# Standard libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import re

# Personal library - Contains all the functions for this notebook
try:
    import myfunctions as mf
except ImportError:
    print("Could not import 'myfunctions.py'. Make sure it exists in the same folder as this notebook.")

# Configure logger
mf.configure_logger()
```

```python
# General path to the data
data_directory = "./"

# List available surfaces
surfaces = sorted([
    item for item in os.listdir(data_directory)
    if os.path.isdir(os.path.join(data_directory, item))
    and item.startswith("Au")
])

print("Available Surfaces:", surfaces)
```

```python
# Upload data
data = mf.load_all_data()

# Display key results
combined = []
for energy_path, energy in data['dft_energies'].items():
    base_folder = energy_path.parent
    freq_path = base_folder / "frequency.dat"
    thermal = data['thermal_corrections'].get(freq_path, None)
    combined.append({
        "path": str(base_folder),
        "energy": energy,
        "thermal_correction": thermal,
    })

df_combined = pd.DataFrame(combined)
df_combined
```
